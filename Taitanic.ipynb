```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("D:\\excel files\\titanic.csv")



print(df.describe())
print(df.info())

df
```

حذف ستون نام ها و جایگزین جنسیت هابا عدد 1 و 2

```python
df.drop(columns= ["Name"] , inplace=True )
#df.drop(columns= ["Cabin"] , inplace=True )
#df.drop(columns= ["Ticket"] , inplace=True )
df["Sex"] = df["Sex"].map({"male":1 , "female" : 2 })
df["Embarked"] = df["Embarked"].map({"S":3 , "C" : 4 , "Q" : 5 })
df
```

تعداد زنان بیشتری (بیش از دو برابر) از حادثه جان سالمبه در بردند

```python

extracted1 = df[df["Survived"] == 1]

plt.hist(extracted1["Sex"])
plt.show()
```

افرادی که بلیط فرست کلاس داشتن بیشتر زنده موندن (و کسایی که کلاس بی بودن بیشتر از کسایی که کلاس سی بودن مردن)

```python
extracted2 = df[df["Survived"] == 1]

plt.hist(extracted2["Pclass"])
plt.show()
```

کسایی که پول  بلیط بیشتری دادن بیشتر زنده موندن

```python
plt.bar(df["Survived"], df["Fare"])
plt.show()
```

```python
extracted3 = df[df["Survived"] == 1]

plt.hist(extracted3["Embarked"])
plt.show()
```

داده های سن که نان بودن رو تبدیل کردم به میانگین سن ها

```python

df["Age"] = df["Age"].fillna(df["Age"].mean(numeric_only=True))

plt.scatter(df["Survived"], df["Age"] , color = "PINK")
plt.xlabel("Survived",color="RED")
plt.ylabel("Age",color="BLUE")
plt.show()
```
